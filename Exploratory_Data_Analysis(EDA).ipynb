## Outliers

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
outliers =[]

```

```python
dataset=[12,15,13,14,16,11,123,12,16,14,15,13,15,122,11,13,14,16,10,12,11,156,14,18,19,18,15,11,10,16]
```

```python
plt.hist(dataset)
```

```python
outliers=[]

def detect_outliers(data):
    threshold=3 ## 3rd S.D.
    mean=np.mean(data)
    std=np.std(data)

    for i in data:
        z_score=(i-mean/std)
        if np.abs(z_score)> threshold:
            outliers.append(i)
    return outliers



```

```python
detect_outliers(dataset)
```

       3. IQR = (Q3-Q1)
       4. find the lower fence ( Q1 - 1.5(IQR))
       5. find the upper fence (Q3 + 1.5(IQR))

IQR = 75th quantile - 25th quantile

X_scaled = (X - X.median) / IQR

```python
df=pd.read_csv(r'C:\Users\user\Desktop\titanic_train.csv',usecols=['Age','Fare','Survived'])
```

```python
### fillnan

df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
import scipy.stats as stat
import pylab
```

```python
### if we want check wheather feature is normally or gaussian distributed
### Q-Q plot

def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'Age')
```

## Logarithmic distribution


```python
import numpy as np
df['Age_log']=np.log(df['Age'])
```

```python
plot_data(df,'Age_log')
```

## Reciprocal dsitribution

```python
df['Age_reciprocal']=1/df.Age
plot_data(df,'Age_reciprocal')
```

## Square root transformation

```python
df['Age_square']=df.Age**1/2
plot_data(df,'Age_square')
```

### Exponential transformation

```python
df['Age_exponential']=df.Age**(1/1.2)
plot_data(df,'Age_exponential')
```
